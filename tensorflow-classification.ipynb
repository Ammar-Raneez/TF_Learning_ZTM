### Improving model

1. Create - increase layers & number of units, change activation function
2. Compiling - a different optimizer, a different learning rate

```python
# Increase layers & units
tf.random.set_seed(42)

model_3 = tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1),
])

model_3.compile(
    loss=tf.keras.losses.BinaryCrossentropy(),
    optimizer=tf.keras.optimizers.Adam(),
    metrics=['accuracy']
)

model_3.fit(X, y, epochs=100)
```

To visualize our model's predictions we're going to create a function which:

* Takes in a model, features (X) and labels (y).
* Creates a meshgrid of the different X values.
* Makes predictions across the meshgrid.
* Plots the predictions as well as a line between the different classes.

```python
# Visualize what is happening
import numpy as np

def plot_decision_boundary(model, X, y):
    # Define the axis boundaries of the plot and create a meshgrid
    x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
    y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1
    xx, yy = np.meshgrid(
        np.linspace(x_min, x_max, 100),
        np.linspace(y_min, y_max, 100)
    )

    # Create X values (we're going to predict on all of these)
    x_in = np.c_[xx.ravel(), yy.ravel()] # stack 2D arrays together: https://numpy.org/devdocs/reference/generated/numpy.c_.html

    # Make predictions using the trained model
    y_pred = model.predict(x_in)

    # Check for multi-class
    if model.output_shape[-1] > 1: # checks the final dimension of the model's output shape, if this is > (greater than) 1, it's multi-class 
        print("doing multiclass classification...")
        # We have to reshape our predictions to get them ready for plotting
        y_pred = np.argmax(y_pred, axis = 1).reshape(xx.shape)
    else:
        print("doing binary classifcation...")
        y_pred = np.round(np.max(y_pred, axis = 1)).reshape(xx.shape)

    # Plot decision boundary
    plt.contourf(xx, yy, y_pred, cmap=plt.cm.RdYlBu, alpha = 0.7)
    plt.scatter(X[:, 0], X[:, 1], c = y, s = 40, cmap = plt.cm.RdYlBu)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
```

```python
plot_decision_boundary(model_3, X, y)
```

```python
# Attempt to use the model in regression
tf.random.set_seed(42)

# Create some regression data
X_regression = np.arange(0, 1000, 5)
y_regression = np.arange(100, 1100, 5)  # y = x + 100

# Split it into training and test sets
X_reg_train = X_regression[:150]
X_reg_test = X_regression[150:]
y_reg_train = y_regression[:150]
y_reg_test = y_regression[150:]

# Expand dims required since the data here is just a single vector
model_3.fit(np.expand_dims(X_reg_train, axis=-1), y_reg_train, epochs=100)
```

*Model cannot be directly fit, since it was compiled on BinaryCrossentropy*

```python
# Create a new model
tf.random.set_seed(42)

model_3_reg = tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1),
])

model_3_reg.compile(
    loss=tf.keras.losses.mae,
    optimizer=tf.keras.optimizers.Adam(),
    metrics=['mae']
)

model_3_reg.fit(np.expand_dims(X_reg_train, axis=-1), y_reg_train, epochs=100)
```

```python
# Make predictions with our trained model
y_reg_preds = model_3_reg.predict(y_reg_test)

# Plot the model's predictions against our regression data
plt.figure(figsize=(10, 7))
plt.scatter(X_reg_train, y_reg_train, c='b', label='Training data')
plt.scatter(X_reg_test, y_reg_test, c='g', label='Testing data')
plt.scatter(X_reg_test, y_reg_preds.squeeze(), c='r', label='Predictions')
plt.legend();
```

In regression this works since the data grows linearly, therefore a linear line of separation is enough.

However, the classification data will never be separated well through just a linear function - therefore, a non-linear function is required (No activation specified in layer defaults to linear activation)

```python
# Change activation function
tf.random.set_seed(42)

model_4 = tf.keras.Sequential([
    tf.keras.layers.Dense(1, activation='linear')
])

model_4.compile(
    loss=tf.keras.losses.BinaryCrossentropy(),
    optimizer=tf.keras.optimizers.Adam(),
    metrics=['accuracy']
)
history = model_4.fit(X, y, epochs=100)
```

```python
plot_decision_boundary(model_4, X, y)
```

```python
tf.random.set_seed(42)

model_5 = tf.keras.Sequential([
    tf.keras.layers.Dense(1, activation='relu')
])

model_5.compile(
    loss=tf.keras.losses.BinaryCrossentropy(),
    optimizer=tf.keras.optimizers.Adam(),
    metrics=['accuracy']
)
history = model_5.fit(X, y, epochs=100)
```

```python
plot_decision_boundary(model_5, X, y)
```

```python
# Single layer of relu didn't work
tf.random.set_seed(42)

model_5 = tf.keras.Sequential([
  tf.keras.layers.Dense(4, activation='relu'),
  tf.keras.layers.Dense(4, activation='relu'),
])

model_5.compile(
    loss=tf.keras.losses.binary_crossentropy,
    optimizer=tf.keras.optimizers.Adam(),
    metrics=['accuracy']
)

history = model_5.fit(X, y, epochs=200)
```

```python
plot_decision_boundary(model_5, X, y)
```

```python
# Change output layer to single neuron since its binary?
tf.random.set_seed(42)

model_6 = tf.keras.Sequential([
    tf.keras.layers.Dense(4, activation='relu'),
    tf.keras.layers.Dense(4, activation='relu'),
    tf.keras.layers.Dense(1)
])

model_6.compile(
    loss=tf.keras.losses.binary_crossentropy,
    optimizer=tf.keras.optimizers.Adam(),
    metrics=['accuracy']
)

history = model_6.fit(X, y, epochs=200)
```

```python
plot_decision_boundary(model_6, X, y)
```

Still no difference?

```python
# Add activation function for output layer
tf.random.set_seed(42)

model_7 = tf.keras.Sequential([
    tf.keras.layers.Dense(4, activation='relu'),
    tf.keras.layers.Dense(4, activation='relu'),
    
    # Sigmoid is tailored for binary classification (converts output to a value 0-1)
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model_7.compile(
    loss=tf.keras.losses.binary_crossentropy,
    optimizer=tf.keras.optimizers.Adam(),
    metrics=['accuracy']
)

history = model_7.fit(X, y, epochs=200)
```

```python
plot_decision_boundary(model_7, X, y)
```

Finally.
A straight line simply could not be fit in the above dataset. Therefore, all it required was a non-linear function to be almost perfect

**Note**: A combination of Linear & Non-linear functions is one of the key fundamentals of neural networks.

**Side-note**
You can't use ReLU function as the output function for classification tasks because, its range can't represent probability 0 to 1, hence the model doesn't get optimized. That's why it is used only for regression tasks and hidden layers. For binary classification, you have to use an output function with range between 0 to 1 such as sigmoid

### Insight of Activation Functions

```python
A = tf.cast(tf.range(-10, 10), tf.float32)
A
```

```python
plt.plot(A);
```

```python
# Sigmoid function - returns a value between 0 and 1
def sigmoid(X):
    return 1 / (1 + np.exp(-X))
```

```python
sigmoid(A)
```

```python
plt.plot(sigmoid(A));
```

```python
# reLU function - converts negative values to 0
def relu(X):
    return tf.maximum(X, 0)
```

```python
relu(X)
```

```python
plt.plot(relu(A));
```

```python
# Linear function - doesn't do anything (No activation)
tf.keras.activations.linear(A)
```

```python
plt.plot(tf.keras.activations.linear(A));
```

**Side-note**
The model doesn't really learn anything when using only linear activation 
functions, because the linear activation function doesn't change the input data in anyway.

Where as, with non-linear functions, the data gets manipulated. A neural network uses these kind of transformations at a large scale to draw patterns between its inputs and outputs.
